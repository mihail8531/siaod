Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
from __future__ import annotations
from typing import Any
from copy import deepcopy

class Node:
    def __init__(self, value: Any, left: Node=None, right: Node=None):
        self.value = value
        self.left = left
        self.right = right
    
    def __str__(self) -> str:
        return str(self.value)
```

```python
class Deck:
    def __init__(self):
        self.left = None
        self.right = None
        
    def is_empty(self) -> bool:
        return self.right is None
    
    def push_left(self, value: Any) -> Any:
        if self.is_empty():
            node = Node(value)
            self.left = node
            self.right = node
        else:
            node = Node(value, right=self.left)
            self.left.left = node
            self.left = node
        return value
            
    def push_right(self, value: Any) -> Any:
        if self.is_empty():
            node = Node(value)
            self.left = node
            self.right = node
        else:
            node = Node(value, left=self.right)
            self.right.right = node
            self.right = node
        return value
    
    def pop_left(self) -> Any:
        assert not self.is_empty(), "Deck is empty"
        value = deepcopy(self.left.value)
        if self.left.right is None:
            self.left = None
            self.right = None
        else:
            self.left = self.left.right
            self.left.left = None
        return value
    
    def pop_right(self) -> Any:
        assert not self.is_empty(), "Deck is empty"
        value = deepcopy(self.right.value)
        if self.right.left is None:
            self.right = None
            self.left = None
        else:
            self.right = self.right.left
            self.right.right = None
        return value
    
    def peek_left(self) -> Any:
        assert not self.is_empty(), "Deck is empty"
        return self.left.value
    
    def peek_right(self) -> Any:
        assert not self.is_empty(), "Deck is empty"
        return self.right.value
    
    def __str__(self) -> str:
        curr_node = self.left
        res_list: list[str] = []
        while curr_node is not None:
            res_list.append(str(curr_node))
            curr_node = curr_node.right
        return ' '.join(res_list) 
```

```python
from typing import Callable
class Stack(Deck):
    def __init__(self):
        super().__init__()
        self.push = self.push_right
        self.pop = self._is_empty_decorator(self.pop_right)
        self.peek = self._is_empty_decorator(self.peek_right)
        
    def push_left(self, value: Any) -> Any:
        raise AttributeError()
    
    def pop_left(self) -> Any:
        raise AttributeError()
    
    def peek_left(self) -> Any:
        raise AttributeError()
        
    def _is_empty_decorator(self, func: Callable[[], Any]) -> Callable[[], Any]:
        def wrapper():
            assert not self.is_empty(), "Stack is empty"
            return func()
        return wrapper
```

### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
def deck_sort(deck: Deck):
    tmp_deck = Deck()
    while True:
        while True:
            if deck.is_empty():
                while not tmp_deck.is_empty():
                    deck.push_left(tmp_deck.pop_right())
                return deck
            if tmp_deck.is_empty():
                if deck.peek_left() < deck.peek_right():
                    tmp_deck.push_right(deck.pop_left())
                else:
                    tmp_deck.push_right(deck.pop_right())
            else:
                if deck.peek_left() < deck.peek_right():
                    if deck.peek_left() >= tmp_deck.peek_right():
                        tmp_deck.push_right(deck.pop_left())
                    elif deck.peek_right() >= tmp_deck.peek_right():
                        tmp_deck.push_right(deck.pop_right())
                    elif deck.peek_right() <= tmp_deck.peek_left():
                        tmp_deck.push_left(deck.pop_right())
                    elif deck.peek_left() <= tmp_deck.peek_left():
                        tmp_deck.push_left(deck.pop_left())
                    else:
                        break
                else:
                    if deck.peek_right() >= tmp_deck.peek_right():
                        tmp_deck.push_right(deck.pop_right())
                    elif deck.peek_left() >= tmp_deck.peek_right():
                        tmp_deck.push_right(deck.pop_left())
                    elif deck.peek_left() <= tmp_deck.peek_left():
                        tmp_deck.push_left(deck.pop_left())
                    elif deck.peek_right() <= tmp_deck.peek_left():
                        tmp_deck.push_left(deck.pop_right())
                    else:
                        break
        while not tmp_deck.is_empty():
            deck.push_left(tmp_deck.pop_right())
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
class Tower(Stack):
    def __init__(self, name: str):
        self.name = name
        super().__init__()
    
def hanoi(a: Tower, b: Tower, c: Tower, n: int):
    if n == 0:
        return 
    hanoi(a, c, b, n - 1)
    c.push(a.pop())
    for tower in sorted([a, b, c], key=lambda x: x.name):
        print(tower.name, tower)
    print('')
    hanoi(b, a, c, n - 1)

def task3():
    print("input:", end="\n\n")
    with open("lab4_io_files/task3_input.txt", 'r') as f:
        n = int(f.readline().strip())
        print(f"n = {n}")
        a = Tower("a")
        b = Tower("b")
        c = Tower("c")
        disks = sorted(map(int, f.readline().strip().split()), reverse=True)
        print("disks:")
        print(*disks)
        for disk in disks:
            a.push(disk)
        print("\noutput:", end="\n\n")
        hanoi(a, b, c, n)
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
def balance4() -> bool:
    stack = Stack()
    print("input:",end="\n\n")
    with open("lab4_io_files/task4_input.txt", 'r') as f:
        while (char := f.read(1)) != "":
            print(char, end='')
            if char == "":
                break
            if char == "(":
                stack.push(char)
            elif char == ")":
                if stack.is_empty():
                    return False
                stack.pop()
    return stack.is_empty()

def task4():
    res = balance4()
    print("\noutput:", end='\n\n')
    print("ok" if res else "not ok")
```

### Задание №6

```python
def task6():
    stack_alpha = Stack()
    stack_other = Stack()
    print("input:", end='\n\n')
    with open("lab4_io_files/task6_input.txt", 'r') as f:
        print(f.read()) # чтение 2 раза только для того, чтобы показать ввод (необязательно)
        f.seek(0)
        print("\noutput:", end='\n\n')
        while (char := f.read(1)) != "":
            if char.isdigit():
                print(char, end='')
            elif char.isalpha():
                stack_alpha.push(char)
            else:
                stack_other.push(char)
        tmp = 0
        while not stack_alpha.is_empty():
            stack_other.push(stack_alpha.pop())
            tmp += 1
        for i in range(tmp):
            print(stack_other.pop(), end='')
        while not stack_other.is_empty():
            stack_alpha.push(stack_other.pop())
        while not stack_alpha.is_empty():
            print(stack_alpha.pop(), end='')
```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.

```python
def task7():
    deck = Deck()
    print("input:", end='\n\n')
    with open("lab4_io_files/task7_input.txt", 'r') as f:
        for line in f:
            line = line.strip()
            print(line)
            for number in map(int, line.split()):
                if number < 0:
                    deck.push_left(number)
                elif number > 0:
                    deck.push_right(number)
    while not deck.is_empty():
        number = deck.pop_left()
        if number > 0:
            deck.push_left(number)
            break
        deck.push_right(number)
    print("\noutput:", end='\n\n')
    while not deck.is_empty():
        number = deck.pop_right()
        if number > 0:
            deck.push_right(number)
            break
        print(number, end=' ')
    while not deck.is_empty():
        print(deck.pop_left(), end=' ')
            
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
def task8():
    stack = Stack()
    print("input:", end='\n\n')
    with open("lab4_io_files/task8_input.txt", 'r') as f:
        for line in f:
            print(line, end='')
            stack.push(line)
    print("\noutput:", end='\n\n')
    with open("lab4_io_files/task8_output.txt", 'w') as f:
        while not stack.is_empty():
            line = stack.pop()
            print(line, end='')
            f.write(line)
        
```
