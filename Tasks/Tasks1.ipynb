# Задание 1

Даны две строки: `s1` и `s2` с одинаковым размером.

Проверьте, может ли некоторая перестановка строки `s1` “победить” некоторую перестановку строки `s2` или наоборот.

Строка x может “победить” строку y (обе имеют размер n), если `x[i] >= y[i]` (в алфавитном порядке) для всех i от 0 до n-1.


## Ввод

`abc`
`xya`

## Вывод

`True`

### Исполняемый код записать в ячейку ниже:

```python
def task1(s1: str, s2: str) -> bool:
    return any(map(all, zip(*map(lambda x: (x[0] >= x[1], x[1] >= x[0]), zip(sorted(s1), sorted(s2))))))
```

# Задание 2

Дана строка `s`, вернуть самую длинную полиндромную подстроку в `s`.

## Ввод

`babad`

## Вывод

`aba` или `bab`

## Исполняемый код записывать в ячейку ниже

```python
def manacher_odd(s: str) -> list[int]:
    n = len(s)
    d = [0] * n
    l, r = 0, -1
    for i in range(n):
        k = 1 if i > r else min(d[l + r - i], r - i + 1)
        while (l_c := i - k) >= 0 and (r_c := i + k) < n and s[l_c] == s[r_c]:
            k += 1
        d[i] = k
        if i + k - 1 > r:
            l, r = i - k + 1, i + k - 1
    return d      
    
```

```python
def task2(s: str) -> str:
    if len(s) < 2:
        return s
    d1 = list(map(lambda x: x * 2 - 1, manacher_odd(s)))
    d2 = list(map(lambda x: x // 2 * 2, manacher_odd('|'.join(s))[1::2]))
    max_pal = lambda x, k: max(zip(x, range(len(x)), [k] * len(x)))
    d, i, k = max(max_pal(d1, 0), max_pal(d2, 1))
    return s[i - d // 2 + k: i + d // 2 + 1]
```

# Задание 3

Вернуть количество отдельных непустых подстрок текста, которые могут быть записаны как конкатенация некоторой строки с самой собой (т.е. она может быть записана, как `a + a`, где `a` - некоторая строка).

## Ввод

`aabb`

## Вывод

2

## Исполняемый код записывать в ячейку ниже

```python
k = 129
a = ord('a')
def calc_hashes(s: str) -> tuple[list[int], list[int]]:
    n = len(s)
    assert n > 1, "String too short"
    k_pows = [1]
    res = [0, (ord(s[0]) - a + 1)]
    for c in s[1:]:
        k_pows.append(k_pows[-1] * k)
        res.append(res[-1] + (ord(c) - a + 1) * k_pows[-1])
    return (res, k_pows)
```

```python
def task3(s: str) -> int:
    n = len(s)
    d: dict[str, list[int, list[int]]] = dict()
    for i, c in enumerate(s):
        if c in d:
            d[c][1].append(i)
        else:
            d[c] = [1, [i]]
    hashes, k_pows = calc_hashes(s)
    res = 0
    for i, c in enumerate(s):
        for j in d[c][1][d[c][0]:]:
            rad = j - i
            if rad == 1:
                res += 1
                continue
            if j + rad > n:
                break
            if ((hashes[j] - hashes[i]) // k_pows[i] ==
                (hashes[j + rad] - hashes[j]) // k_pows[j]):
                res += 1
                continue
        d[c][0] += 1
    return res
```
