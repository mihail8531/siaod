```python
from typing import List
# 1. Сортировка выбором
def selection_sort(x:List[float]) -> List[float]:
    n = len(x)
    for i in range(n - 1):
        m, m_i = x[i], i
        for j in range(i + 1, n):
            if x[j] < m:
                m, m_i = x[j], j
        x[i], x[m_i] = x[m_i], x[i]
    return x
```

```python
# 2. Сортировка вставкой
def insertion_sort(x:List[float]) -> List[float]:
    n = len(x)
    for i in range(min(1, n-1), n):
        k = x[i]
        j = 0
        while j <= i:
            if k <= x[j]:
                x.insert(j, k)
                break
            j += 1
        x.pop(i + 1)
    return x
#x = [random.randint(-100, 100) for i in range(20)]
#s_s = selection_sort(x.copy())
#i_s = insertion_sort(x.copy())
#print(s_s)
#print(i_s)
#print(all(map(lambda k: k[0] == k[1], zip(s_s,i_s))))
```

```python
# 3. Сортировка обменом
def bubble_sort(x:List[float]) -> List[float]:
    n = len(x)
    for i in range(n - 1):
        for j in range(0, n - 1 - i):
            if x[j] >= x[j + 1]:
                x[j], x[j + 1] = x[j + 1], x[j]
    return x 
```

```python
# 6.Быстрая сортировка
def quick_sort(x:List[float]) -> List[float]:
    n = len(x)
    if (n < 2):
        return x
    l, m, r = [], [], []
    for i in range(0, n):
        if x[i] < x[0]:
            l.append(x[i])
        elif x[i] == x[0]:
            m.append(x[i])
        else:
            r.append(x[i])
    return quick_sort(l) + m + quick_sort(r)

x = [random.randint(-100, 100) for i in range(20)]
s_s = selection_sort(x.copy())
q_s = quick_sort(x.copy())
print(q_s)
print(all(map(lambda k: k[0] == k[1], zip(s_s, q_s))))
```

```python
#7 Пирамидальная сортировка
def heapify(x:List[float], n:int, i:float):
    p_i = i
    l_i = p_i * 2 + 1
    r_i = l_i + 1
    if l_i < n and x[l_i] > x[p_i]:
        p_i = l_i
    if r_i < n and x[r_i] > x[p_i]:
        p_i = r_i
    if p_i != i:
        x[p_i], x[i] = x[i], x[p_i]
        heapify(x, n, p_i)
```

```python
def heap_sort(x:List[float]) -> List[float]:
    n = len(x)
    if (n < 2):
        return x
    for i in range(n // 2 + 1, -1, -1):
        heapify(x, n, i)
    for i in range(n - 1, 0, -1):
        x[0], x[i] = x[i], x[0]
        heapify(x, i, 0)
    return x

x = [random.randint(-100, 100) for i in range(20)]
s_s = selection_sort(x.copy())
h_s = heap_sort(x.copy())
print(h_s)
print(s_s)
print(all(map(lambda k: k[0] == k[1], zip(s_s, h_s))))
```
