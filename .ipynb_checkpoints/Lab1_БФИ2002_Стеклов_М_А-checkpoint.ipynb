```python
from typing import List
# 1. Сортировка выбором
def selection_sort(x:List[float]) -> List[float]:
    n = len(x)
    if n < 2:
        return x
    for i in range(n - 1):
        m, m_i = x[i], i
        for j in range(i + 1, n):
            if x[j] < m:
                m, m_i = x[j], j
        x[i], x[m_i] = x[m_i], x[i]
    return x
```

```python
# 2. Сортировка вставкой
def insertion_sort(x:List[float]) -> List[float]:
    n = len(x)
    if n < 2:
        return x
    for i in range(min(1, n-1), n):
        k = x[i]
        j = 0
        while j <= i:
            if k <= x[j]:
                x.insert(j, k)
                break
            j += 1
        x.pop(i + 1)
    return x
#x = [random.randint(-100, 100) for i in range(20)]
#s_s = selection_sort(x.copy())
#i_s = insertion_sort(x.copy())
#print(s_s)
#print(i_s)
#print(all(map(lambda k: k[0] == k[1], zip(s_s,i_s))))
```

```python
# 3. Сортировка обменом
def bubble_sort(x:List[float]) -> List[float]:
    n = len(x)
    if n < 2:
        return x
    for i in range(n - 1):
        for j in range(0, n - 1 - i):
            if x[j] >= x[j + 1]:
                x[j], x[j + 1] = x[j + 1], x[j]
    return x 
```

```python
#5. Турнирная сортировка
def min_heapify(x:List[float], n:int, i:int):
    p_i = i
    l_i = p_i * 2 + 1
    r_i = l_i + 1
    if l_i < n and x[l_i] < x[p_i]:
        p_i = l_i
    if r_i < n and x[r_i] < x[p_i]:
        p_i = r_i
    if p_i != i:
        x[p_i], x[i] = x[i], x[p_i]
        min_heapify(x, n, p_i)
```

```python
def fill_tournament(x:List[float], n:int, tournament:List[float], leafs:int, i:int, losers:List[float], winners:List[float]) -> int:
    for j in range(-leafs, 0, 1):
        if math.isinf(tournament[j]):
            if len(winners) > 0 and x[i] < winners[-1]:
                losers.append(x[i])
            else:
                tournament[j] = x[i]
            i += 1
        if i >= n:
            return i
    return i
    
```

```python
def merge(x:List[float], y:List[float]):
    res = []
    i, j = 0, 0
    while i < len(x) and j < len(y):
        if x[i] < y[j]:
            res.append(x[i])
            i += 1
        else:
            res.append(y[j])
            j += 1
    return res + x[i:] + y[j:]
```

```python
def tournament_sort(x:List[float], layers=3) -> List[float]:
    n = len(x)
    if n < 2:
        return x
    losers, winners = [], []
    leafs = 2 ** (layers - 1)
    tournament_list = [math.inf] * (2 ** layers - 1)
    j = fill_tournament(x, n, tournament_list, leafs, 0, losers, winners)
    for i in range(n // 2 + 1, -1, -1):
        min_heapify(tournament_list, len(tournament_list), i)
    winners.append(tournament_list[0])
    tournament_list[0] = math.inf
    while (len(losers) + len(winners) < n):
        if (j < n):
            j = fill_tournament(x, n, tournament_list, leafs, j, losers, winners)
        for i in range(n // 2 + 1, -1, -1):
            min_heapify(tournament_list, len(tournament_list), i)
        winners.append(tournament_list[0])
        tournament_list[0] = math.inf
    x = losers + winners
    if len(losers) > 0:
        # x = tournament_sort(x)
        losers = tournament_sort(losers)
        x = merge(losers, winners)
    return x
for i in range(1000):
    x = [random.randint(-100, 100) for i in range(20)]
    t_s = tournament_sort(x.copy())
```

```python
# 6.Быстрая сортировка
def quick_sort(x:List[float]) -> List[float]:
    n = len(x)
    if (n < 2):
        return x
    l, m, r = [], [], []
    for i in range(0, n):
        if x[i] < x[0]:
            l.append(x[i])
        elif x[i] == x[0]:
            m.append(x[i])
        else:
            r.append(x[i])
    return quick_sort(l) + m + quick_sort(r)
```

```python
#7 Пирамидальная сортировка
def heapify(x:List[float], n:int, i:float):
    p_i = i
    l_i = p_i * 2 + 1
    r_i = l_i + 1
    if l_i < n and x[l_i] > x[p_i]:
        p_i = l_i
    if r_i < n and x[r_i] > x[p_i]:
        p_i = r_i
    if p_i != i:
        x[p_i], x[i] = x[i], x[p_i]
        heapify(x, n, p_i)
```

```python
def heap_sort(x:List[float]) -> List[float]:
    n = len(x)
    if (n < 2):
        return x
    for i in range(n // 2 + 1, -1, -1):
        heapify(x, n, i)
    for i in range(n - 1, 0, -1):
        x[0], x[i] = x[i], x[0]
        heapify(x, i, 0)
    return x

x = [random.randint(-100, 100) for i in range(20)]
s_s = selection_sort(x.copy())
h_s = heap_sort(x.copy())
print(h_s)
print(s_s)
print(all(map(lambda k: k[0] == k[1], zip(s_s, h_s))))
```
