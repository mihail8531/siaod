–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ k (0‚Äâ‚â§‚Äâk‚Äâ<‚Äâ360). –í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∞—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ x –∏ y (‚Äâ-‚Äâ1390‚Äâ‚â§‚Äâx,‚Äây‚Äâ‚â§‚Äâ1390).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í—ã–≤–µ–¥–∏—Ç–µ –¥–≤–∞ —á–∏—Å–ª–∞ ‚Äî —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—Ä–∞—â–µ–Ω–∏—è. –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –≤–∞—à–µ–≥–æ –æ—Ç–≤–µ—Ç–∞ –Ω–µ –¥–æ–ª–∂–Ω–∞ –ø—Ä–µ–≤—ã—à–∞—Ç—å 10‚Äâ-‚Äâ1.

–ü—Ä–∏–º–µ—Ä—ã
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
90

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
-1.00000000 1.00000000
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
180
1 1390
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
-1.00000000 -1390.00000000

```python
from math import sqrt, cos, sin, pi

def task1(k: int, x: int, y: int) -> tuple[float, float]:
    """
    :param k: Degree by which the point will rotate
    :param x: X coordinate of the point
    :param y: Y coordinate of the point
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: (x2, y2) where x2, y2 coordinates of the result point
    """
    if not (isinstance(k, int) and isinstance(x, int) and isinstance(y, int)):
        raise TypeError("k, x and y must be integers")
    if not (0 <= k < 360):
        raise ValueError("k must be in range from 0 to 359 inclusive")
    if not ((-1390 <= x <= 1390) and (-1390 <= y <= 1390)):
        raise ValueError("x and y must be in range from -1390 to 1390 inclusive")
    angle = pi * k / 180
    x2 = x * cos(angle) - y * sin(angle)
    y2 = x * sin(angle) + y * cos(angle)
    return x2, y2
    
```

```python
k = int(input())
x, y = map(int, input())
x2, y2 = task1(k, x, y)
print(x2, y2)
```

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è —á–∏—Å–ª–æ n (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ1000), n ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –≤ –≥—Ä—É–ø–ø–µ. –í —Å–ª–µ–¥—É—é—â–µ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∞—Ç—Å—è —ç–ª–µ–º–µ–Ω—Ç—ã, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –≤ –ø–æ—Ä—è–¥–∫–µ –Ω–µ—É–±—ã–≤–∞–Ω–∏—è. –≠—Ç–∏ —á–∏—Å–ª–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –∑–∞–ø—è—Ç–æ–π, —Å–æ–¥–µ—Ä–∂–∞—Ç –Ω–µ –±–æ–ª—å—à–µ —Å–µ–º–∏ —Ü–∏—Ñ—Ä –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ —Å—á–∏—Å–ª–µ–Ω–∏—è –∏ –Ω–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤–µ–¥—É—â–∏—Ö –Ω—É–ª–µ–π.\
–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–ï—Å–ª–∏ –≥—Ä—É–ø–ø–∞ —á–∏—Å–µ–ª —è–≤–ª—è–µ—Ç—Å—è –≥—Ä—É–ø–ø–æ–π –¥—Ä—É–∑–µ–π, –≤—ã–≤–µ–¥–∏—Ç–µ ¬´FRIENDS¬ª, –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–µ–¥–∏—Ç–µ ¬´NOT FRIENDS¬ª.\
–ü—Ä–∏–º–µ—Ä:\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
4\
1,3,6,12\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
FRIENDS\

```python
def task2(n: int, a: list[int]) -> str:
    """
    :param n: Lenght of the list
    :param a: List of integers
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: "FRIENDS" or "NOT FRIENDS"
    """
    if not isinstance(n, int):
        raise TypeError("n must be an integer")
    if not isinstance(a, list) or any(not isinstance(x, int) for x in a):
        raise TypeError("a must be a list of integers")
    if not (1 <= n <= 1000):
        raise ValueError("n must be in range from 1 to 1000 inclusive")
    if len(a) != n:
        raise ValueError("n must be equal to lenght of the list a")
    if any(abs(x) >= 10000000 or x == 0 for x in a):
        raise ValueError("absolute value of a list element must be smaller than \
                         10000000 and can't be a zero")
    if any(a[i] > a[i + 1] for i in range(n - 1)):
        raise ValueError("array must be sorted")
    for i in range(n):
        if a[i] > 0:
            break
    a = list(map(abs, a))
    a = a[i::-1][:-1] + a[i:]
    sorted_abs = []
    l, r = 0, i
    while l < i and r < n:
        if a[l] < a[r]:
            sorted_abs.append(a[l])
            l += 1
        else:
            sorted_abs.append(a[r])
            r += 1
    if l < i:
        sorted_abs.extend(a[l:i])
    else:
        sorted_abs.extend(a[r:n])
    if all(a[i+1] % a[i] == 0 for i in range(n - 1)):
        return "FRIENDS"
    return "NOT FRIENDS"
```

```python
n = int(input())
a = list(map(int, input().split(',')))
res = task2(n, a)
print(res)
```

```python
from string import ascii_letters
def task3(n: int, s_1: str, s_2: str) -> str:
    """
    :param n: Friends amount
    :param s_1: Initial string
    :param s_2: Result string
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: "Yes" or "No"
    """
    if not isinstance(n, int):
        raise TypeError("n must be an integer")
    if not (isinstance(s_1, str) and isinstance(s_2, str)): 
        raise TypeError("s_1 and s_2 must be strings")
    if not (2 <= n <= 8 * 10**6):
        raise ValueError("n must be in range from 2 to 8 * 10**6 inclusive")
    if len(s_1) > 10**7:
        raise ValueError("s_1 lenght must be smaller or equal than 10**7")
    if not (all(x in ascii_letters for x in s_1) and all(x in ascii_letters for x in s_2)):
        raise ValueError("s_1 and s_2 must contain only english letters")
    for i in range(min(len(s_1), len(s_2))):
        if s_1[i] != s_2[i]:
            break
    if len(s_2) - i + len(s_1) - i <= n:
        return "Yes"
    return "No"
```

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –æ–¥–Ω–æ –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ10^8).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
upset

```python
def task4(str_n: str) -> str:
    """
    :param str_n: Number as string
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: "heppy",  "heppier" or "upset"
    """
    if not isinstance(str_n, str):
        raise TypeError("str_n must be a string")
    n = int(str_n) #raises ValueError if str_n is bad
    if not (1 <= n <= 10**8):
        raise ValueError("n must be in range from 1 to 10**8 inclusive")
    checked_digits = (n % int(x) == 0 for x in str_n)
    if all(checked_digits):
        return "heppier"
    elif any(checked_digits):
        return "happy"
    else:
        return "upset"
```

```python
str_n = input()
res = task4(str_n)
print(res)
```

```python
def task5(a: int, b: int, c: int) -> int:
    """
    :param a: Interger a
    :param b: Interger b
    :param c: Interger c
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: a**b % c
    """
    if not (isinstance(a, int) or isinstance(b, int) or isinstance(c, int)):
        raise TypeError("a, b and c must be integers")
    if not (1 <= a <= 100 and 1 <= b <= 100 and 1 <= c <= 100):
        raise ValueError("a, b and c must be in range from 1 in 100 inclusive")
    return a**b % c
    
```

```python
def task8(s: str, p: float) -> float:
    """
    :param s: Initial string
    :param p: Probabilty of replacing '?' with '1'
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range 
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: Expected value
    """
    available_chars = ('?', '0', '1')
    if not isinstance(s, str):
        raise TypeError("s must be a string")
    if not isinstance(p, float):
        raise TypeError("p must be a float")
    if any(c not in available_chars for c in s):
        raise ValueError("s must contain only '?', '0' and '1'")
    clear_s = s.replace('?','')
    ones = sum(map(int, clear_s))
    questions = len(s) - len(clear_s)
    return round((p*questions + ones) / len(s), 5)
```

```python
RootsList = list[tuple[int, int]]

def task9(a: int, b: int, c: int) -> tuple[int, RootsList]:
    """
    :param a: Coefficient a
    :param b: Coefficient c
    :param c: Coefficient b
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range 
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: (n, [(x1, y1), (x2, y2), ... , (xn, yn)]) where xi, yi roots of A*X‚Äâ+‚ÄâB*Y‚Äâ=‚ÄâC
    """
    if not (isinstance(a, int) and isinstance(b, int) and isinstance(c, int)):
        raise TypeError("a, b and c must be integers")
    if not (1 <= a <= 1000 and 1 <= b <= 1000 and 1 <= c <= 1000):
        raise ValueError("a, b and c must be in range from 1 in 1000 inclusive")
    count = 0
    roots = []
    for x in range(1, 1000):
        k = (c - a*x)
        y = k // b
        if k % b == 0:
            roots.append((x, y))
        elif y <= 0:
            break
    return len(roots), roots
        
```

```python
from math import sqrt
from typing import Union

def task10(a: int, b: int, c: int) -> Union[tuple[float], tuple[float, float]]:
    """
    :param a: Coefficient a
    :param b: Coefficient c
    :param c: Coefficient b
    :raises: :class:`ValueError`: Raises if argument is not in the acceptable range 
    :raises: :class:`TypeError`: Raises if argument have wrong type

    :return: return (x1,) or (x1, x2), where x1, x2 roots of the quadratic equation with given coefficients
    """
    if not (isinstance(a, int) and isinstance(b, int) and isinstance(c, int)):
        raise TypeError("a, b and c must be integers")
    if not (-1000 <= a <= 1000 and -1000 <= b <= 1000 and -1000 <= c <= 1000):
        raise ValueError("a, b and c must be in range from -1000 in 1000 inclusive")
    if a == 0:
        raise ValueError("a can't be equal to the zero")
    d = b**2 - 4*a*c
    if d < 0:
        raise ValueError("equation has no real solutions")
    x1, x2 = (-b - sqrt(d)) / (2*a), (-b + sqrt(d)) / (2*a)
    if d == 0:
        return x1,
    return min(x1, x2), max(x1, x2)
```
